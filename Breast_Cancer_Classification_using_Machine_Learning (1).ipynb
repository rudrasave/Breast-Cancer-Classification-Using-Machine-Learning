```python
# ========================================
# Breast Cancer Classification (Hybrid)
# ========================================

# Import libraries
import os
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# ================================
# Load the Breast Cancer Dataset
# ================================
csv_path = "breast_cancer_dataset_augmented.csv"

if os.path.exists(csv_path):
    print(f"📁 Custom dataset found: {csv_path}")
    data_frame = pd.read_csv(csv_path)
    target_col = 'target' if 'target' in data_frame.columns else 'label'
else:
    print("📦 CSV not found. Using sklearn built-in dataset.")
    breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
    data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
    data_frame['label'] = breast_cancer_dataset.target
    target_col = 'label'

# Features and target
X = data_frame.drop(columns=target_col, axis=1)
Y = data_frame[target_col]

print(f"✅ Data loaded successfully! Shape: {X.shape}, Target length: {len(Y)}")

# ================================
# Train-test split
# ================================
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2, stratify=Y
)

# ================================
# Feature Scaling
# ================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ================================
# Model: Random Forest Classifier
# ================================
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}

grid_rf = GridSearchCV(
    RandomForestClassifier(random_state=2),
    param_grid_rf,
    cv=5,
    scoring='accuracy'
)
grid_rf.fit(X_train_scaled, Y_train)

best_rf = grid_rf.best_estimator_

# ================================
# Evaluate Performance
# ================================
train_acc_rf = accuracy_score(Y_train, best_rf.predict(X_train_scaled))
test_acc_rf = accuracy_score(Y_test, best_rf.predict(X_test_scaled))
cv_acc_rf = cross_val_score(best_rf, X, Y, cv=5).mean()

print("\n===== 🌲 Random Forest Classifier =====")
print("Best Parameters:", grid_rf.best_params_)
print(f"Train Accuracy: {train_acc_rf:.4f}")
print(f"Test Accuracy:  {test_acc_rf:.4f}")
print(f"CV Accuracy:    {cv_acc_rf:.4f}")
```

```python
# Breast Cancer Classification (Hybrid Optimized)
import os
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# ------------------------------
# 1️⃣ Load Dataset (CSV or Default)
# ------------------------------
csv_path = "breast_cancer_dataset_augmented.csv"

if os.path.exists(csv_path):
    print(f"📁 Custom dataset found: {csv_path}")
    df = pd.read_csv(csv_path)
    X = df.drop('target', axis=1)
    y = df['target']
else:
    print("📦 CSV not found. Using default sklearn breast cancer dataset.")
    data = load_breast_cancer()
    X = pd.DataFrame(data.data, columns=data.feature_names)
    y = data.target

print(f"✅ Data loaded successfully! Shape: {X.shape}, Target length: {len(y)}")

# ------------------------------
# 2️⃣ Split & Scale
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2, stratify=y
)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ------------------------------
# 3️⃣ Define Models
# ------------------------------
log_reg = LogisticRegression(max_iter=2000, random_state=2)
rf = RandomForestClassifier(n_estimators=300, max_depth=8, random_state=2)
xgb = XGBClassifier(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=2
)

# ------------------------------
# 4️⃣ Ensemble Model (Soft Voting)
# ------------------------------
ensemble = VotingClassifier(
    estimators=[('lr', log_reg), ('rf', rf), ('xgb', xgb)],
    voting='soft'
)

# ------------------------------
# 5️⃣ Train & Evaluate
# ------------------------------
ensemble.fit(X_train, y_train)
train_acc = accuracy_score(y_train, ensemble.predict(X_train))
test_acc = accuracy_score(y_test, ensemble.predict(X_test))
cv_acc = cross_val_score(ensemble, X, y, cv=5).mean()

print(f"\n📊 Train Accuracy: {train_acc:.4f}")
print(f"🧪 Test Accuracy:  {test_acc:.4f}")
print(f"🔁 CV Accuracy:    {cv_acc:.4f}")

# ------------------------------
# 6️⃣ Prediction Example
# ------------------------------
sample = np.array([
    13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,
    0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,
    15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259
])
sample_scaled = scaler.transform(sample.reshape(1, -1))
pred = ensemble.predict(sample_scaled)[0]

print("\n🔍 Prediction:", "Benign" if pred == 1 else "Malignant")
```
